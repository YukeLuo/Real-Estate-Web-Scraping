```python
base_url="http://www.century21.com/real-estate/bellevue-wa/LCWABELLEVUE/"
property_url = "https://www.century21.com/property/"
#for i in data_id:
r=requests.get(base_url, headers=headers)
#print(r)
c=r.content
soup=BeautifulSoup(c,"html.parser")
newall=soup.find_all("div",{"class":"infinite-item"})
#print(newall)
l = []
for item in newall:
    d = {}
    try:
        d["Address"] = item.find("div",{"class":"property-address-info"}).text.replace("\n","").strip().split('                        ')[0]
    except:
        pass
    try:
        d["Locality"] = item.find("div",{"class":"property-address-info"}).text.replace("\n","").strip().split('                        ')[1]
    except:
        pass
    try:
        d["Price"] = item.find("a",{"class":"listing-price"}).text.replace("\n","").replace(" ","")
    except:
        pass
    try:
        d["Beds"] = item.find("div",{"class":"property-beds"}).find("strong").text
    except:
        d["Beds"] = None
    try:
        d["Baths"] = item.find("div",{"class":"property-baths"}).find("strong").text
    except:
        d["Baths"] = None
    try:
        d["Half-Baths"] = item.find("div",{"class":"property-half-baths"}).find("strong").text
    except:
        d["Half-Baths"] = None
    try:
        d["Square-ft"] = item.find("div",{"class":"property-sqft"}).find("strong").text
    except:
        d["Square-ft"] = None
    
    direction = '-'.join((d['Address']+d['Locality']).lower().split(' '))
    try:
        data_id = item['data-id']
    except:
        pass
    #for id in data_id:
    #print(property_url+direction+'-'+data_id)
    newr = requests.get(property_url+direction+'-'+data_id, headers=headers)
    newc=newr.content
    newsoup=BeautifulSoup(newc,"html.parser")
    for i in newsoup.select(".property-features"):
        features = i.text.split("\n")
        for j in range(1,len(features)-1):
            feature = features[j].split(": ")[0]
            value = features[j].split(": ")[1]
            d[feature] = value
    l.append(d)
```
